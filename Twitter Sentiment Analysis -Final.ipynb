### 3. Preprocess

In order to have the best model, preprocessing data is very important. In this section, data will be cleaned through removing stopwords and punctuations to reduce unnecessary noise from the dataset. Also **Stemming** and **Lemmatization** will be performed for better results.

##### A. Create Stoplist

```python
#adding unnecessary words in stoplist
stopwords_list = stopwords.words('english')
stopwords_list += list(string.punctuation)
stopwords_list = ['sxsw', 'mention', 'rt', 'link']
```

```python
stopwords_list
```

```python
#removing stopwords from the list
STOPWORDS = set(stopwords_list)
def cleaning_stopwords(tweets):
    return " ".join([word for word in str(tweets).split() if word not in STOPWORDS])
combined['tweets_sw'] = combined['tweets'].apply(lambda tweets: cleaning_stopwords(tweets))
combined['tweets_sw'].head()
```

##### B. Remove Punctuations

```python
#removing punctuations
punct = string.punctuation
def cleaning_punctuations(tweets_sw):
    translator = str.maketrans('', '', punct)
    return tweets_sw.translate(translator)
combined['tweets_cl']= combined['tweets_sw'].apply(lambda x: cleaning_punctuations(x))
combined['tweets_cl'].head()
```

##### C. Create Tokens

```python
#creating token by using tokenizer
tokenizer = RegexpTokenizer(r"([a-zA-Z]+(?:[’'][a-z]+)?)")
combined['tweets_tk'] = combined['tweets_cl'].apply(tokenizer.tokenize)
combined['tweets_tk'].head()
```

```python
#creating column without applying stopwords
combined['no_stops'] = combined['tweets_tk'].apply(lambda x: [w for w in x if w not in stopwords_list])
combined['no_stops'] 
```

##### D. Applying Stemming

```python
#applying stemmer
st = nltk.PorterStemmer()
def stemming_on_text(data):
    text = [st.stem(word) for word in data]
    return data
combined['stem']= combined['no_stops'].apply(lambda x: stemming_on_text(x))
combined['stem'].head()
```

##### E. Applying Lemmatizer

```python
#applying lemmatizer
lm = nltk.WordNetLemmatizer()
def lemmatizer_on_text(data):
    text = [lm.lemmatize(word) for word in data]
    return data
combined['lemma'] = combined['stem'].apply(lambda x: lemmatizer_on_text(x))
combined['lemma'].head()
```

```python
#reapplying stopwords
stopwords_list = stopwords.words('english')
stopwords_list += list(string.punctuation)
stopwords_list += ['mention', 'sxsw',  'rt', 'link']
combined['stops'] = combined['lemma'].apply(lambda x: [w for w in x if w not in stopwords_list])
```

### 4. Analysis

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.05, random_state =4000)
```

```python
#creating countvetorizer 
cv = CountVectorizer()

X_train_vec = cv.fit_transform(X_train)
X_train_vec = pd.DataFrame.sparse.from_spmatrix(X_train_vec)
X_train_vec.columns = sorted(cv.vocabulary_)
X_train_vec.set_index(y_train.index, inplace=True)

#Used from a lecture[NKP Vectorization](https://github.com/flatiron-school/ds-nlp_vectorization-kvm32/blob/main/nlp_vectorization.ipynb)
```

```python
#checking the vectors
X_train_vec
```

```python
#creating predict values for test and evaluate
def evaluate(model):
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred))
```

##### Model 1 - MultinomialNB

```python
#creating first Model with MultinominalNB
mnb = MultinomialNB()
cv = CountVectorizer()
mnb_pipe = make_pipeline(cv, mnb)

mnb_model = mnb_pipe.fit(X_train, y_train)
evaluate(mnb_model)
plot_confusion_matrix(mnb_pipe, X_test, y_test, values_format= '.0%', cmap='binary', normalize='all')
plt.title('Multinomial confusion matrix')
```

▶ My first model by using [**Multinomial Naive Bayes**](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.MultinomialNB.html) classifier seems have fairly good accuracy at 84%. Let's see if other models would be better.
</br>

**Confusion Matrix**</br>
▶ Predicted that 5% words are Negative emotion and it is true.(True Positive)
</br>
▶ Predicted that 79% wordss are Positive emotion and it is true. (True Negative)
</br>
▶ Predicted that 14% words are Nagative emotion but it is false.(False Negative)
</br>
▶ Predicted that 2% words are Positive emotion but it is false. (False Positive)


##### Model 2 - Random Forst

```python
#creating model with Random Forest Classifier
rfc = RandomForestClassifier()
cv = CountVectorizer()

rfc_pipe = make_pipeline(cv, rfc)

rf_model = rfc_pipe.fit(X_train, y_train)
evaluate(rf_model)
plot_confusion_matrix(rfc_pipe, X_test, y_test, values_format= '.0%', cmap='binary', normalize='all')
plt.title('Random Forest confusion matrix')
```

▶ The accuracy increased to 89%.
</br>
▶ It has 0 False Positive.

##### Model 3 - Logistic Regression

```python
#creating last model with Logistic Regression
lr = LogisticRegression()
cv = CountVectorizer()

lr_pipe = make_pipeline(cv, lr)

lr_model = lr_pipe.fit(X_train, y_train)
evaluate(lr_model)
plot_confusion_matrix(lr_pipe, X_test, y_test, values_format= '.0%', cmap='binary', normalize='all')
plt.title('Logistic Regression confusion matrix')
```

▶ The Accuracy went down by 1 % (compare to random forest). This model is second best from 3 models that I have created.
